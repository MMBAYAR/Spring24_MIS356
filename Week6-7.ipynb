##Generating data with some noise, drawing samples, plotting

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#. predetermining plot format features
plt.rcParams['figure.figsize'] = [6,3.875]
plt.rcParams['font.size'] = 12
plt.rcParams['figure.dpi'] = 200
plt.rcParams["axes.spines.right"] = False
plt.rcParams["axes.spines.top"] = False
```

```python
#. Generating the predictos (X)
X = np.linspace( -1.5, 7, 10**5 )

#. Defining a trig function with a slightly positive trend
def f(x):
  return np.sin( (x-2)**2 ) + 0.4*x
#.   This is the regression function, the function that best explains the data

#. Generating responses with noise (mathing number of random additions to the reg.func.)
y = f(X) + np.random.normal(loc = 0, scale = 0.25, size = 10**5)
```

```python
#. Plotting the regression function in red
sns.lineplot(
    #. Values for the horizontal axis:
    x = X,
    #. Values for the vertical axis:
    y = f(X),
    color = 'red',
    label = '$f(x)$'
)

#. Plotting the responses in blue:
sns.scatterplot(
    #. Values for the horizontal axis:
    x = X,
    #. Values for the vertical axis:
    y = y,
    color = 'blue',
    alpha = .05,
    marker = 'x',
    label = '$y$'
)

#. Renaming axes
plt.xlabel( '$x$' )
plt.ylabel( '$y$' )

#. Plot title
plt.title( 'A combination regression function and the observations' )

#. Downloading image as a .png file
```

##Since we have too many points ($10^5$), they overlap and form a cloud wrapping the regression function

##It is best we draw out a representative sample with reasonable size

```python
#  DRAWING A RANDOM SAMPLE OF 50 DATA

#.   Converting a numpy array into a pandas Series
X = pd.Series( X )
y = pd.Series( y )

#    A random sample, the fraction is sample_size / population_size
sample_size = 250
population_size = len(X)

X_sample = X.sample( frac = sample_size / population_size )

#.   Using the indices of X_sample, defining the correspongind responses
y_sample = y.iloc[ list(X_sample.index) ]
```

```python
#. Plotting the responses again...
#.   Plotting the regression function in red
sns.lineplot(
    #. Values for the horizontal axis:
    x = X,
    #. Values for the vertical axis:
    y = f(X),
    color = 'red',
    alpha = .1,
    label = '$f(x)$'
)

#.   Plotting the responses in blue:
sns.scatterplot(
    #. Values for the horizontal axis:
    x = X_sample,
    #. Values for the vertical axis:
    y = y_sample,
    color = 'blue',
    marker = 'x',
    label = '$y_{sample}$'
)

#.   Plotting deviations from the reg. function
for i in X_sample.index:
  sns.lineplot(
  x = [X_sample[i],X_sample[i]],
  y = [y_sample[i],f( X_sample[i] )],
  color = 'blue',

  )

#. Renaming axes
plt.xlabel( '$x_{sample}$' )
plt.ylabel( '$y_{sample}$' )

#. Plot title
plt.title( 'A combination regression function and the observations' )
```

##. Simple linear regression for predicting y values

#Now lets make this more flexible

```python
#. To make this model more flexible I have exract some polynomial (non-linear) features of the data
#.   maybe I can just catch the trig curves?
from sklearn.preprocessing import PolynomialFeatures

Model_perf = {}

for i in range(1,250):

  poly = PolynomialFeatures(
                            degree = i,
                            include_bias = True,
                            ).fit_transform(
                                            np.array(X_sample).reshape(-1,1)
                                            )

  #. a new model with three features (non-linear) and a single response
  model = LinearRegression().fit(
                                poly, #previously it was : np.array(X_sample).reshape(-1,1),
                                y_sample
                                )

  #. How does my model perform?
  #.    Training set
  traning_mse = mse(
  y_sample,
  model.predict( poly )
  )

  #.   Test set
  test_mse = mse(
  y_test,
  model.predict(
  PolynomialFeatures(
                            degree = i,
                            include_bias = True,
                            ).fit_transform(
                                            np.array(X_test).reshape(-1,1)
                                            )
  )
  )

  Model_perf[i] = (traning_mse,test_mse)

seri = pd.Series([Model_perf[i][1] for i in range(1,50+1)])

seri[seri == seri.min()].index[0] + 1
```

#. Episode 2: Using a non-parametric model = K-Nearest Neighbor Regressor

```python
sns.scatterplot(
    x = X,
    y = y)
```

```python
#. Getting some number of points in a sample
sample_size = 250
X_train = X.sample( frac = sample_size / len(X) )
X_train
```

```python
X_train.index
```

```python
y_train = y.loc[X_train.index]
y_train
```

```python
#  Draw out a test set of size = 200
#.   Eliminate the observations in the training set
X_test = X.loc[ [i for i in X.index if i not in X_train.index] ].sample( frac = 200 / len( X.loc[ [i for i in X.index if i not in X_train.index] ] )  )

y_test = y.loc[ X_test.index ]
y_test
```

```python
#. plotting the training set
sns.scatterplot(
    x = X_train,
    y = y_train,
    label = 'training_set',
    alpha = 0.5
)
#. plotting the set set
sns.scatterplot(
    x = X_test,
    y = y_test,
    label = 'test_set',
    alpha = 0.5
)

plt.xlabel( 'Predictors' )
plt.ylabel( 'Response' )
```

```python
#. Importing the KNN Regression function within the sklearn.neighbors submodule
from sklearn.neighbors import KNeighborsRegressor
```

```python
type( X_train )
```

```python
model = KNeighborsRegressor( n_neighbors = len( X_train) ).fit(
                                                              np.array(X_train).reshape(-1, 1),
                                                              y_train
                                                              )
y_pred = model.predict( np.array(X_test).reshape(-1, 1) )

#. Plot the test set performance of the KNN regressor
sns.scatterplot(
x = X_test,
y = y_test,
label = 'Original test set responses'
)
sns.scatterplot(
x = X_test,
y = y_pred,
label = 'KNN predictions'
)
```

```python
#. Finding the best K-decisiton

#.   to record the mean squared error values in the loop
MSE = {}
#.   Form a loop that tries a number K:
for K in range(1, len(X_train)+1):

  model = KNeighborsRegressor( n_neighbors = K ).fit(
                                                    np.array(X_train).reshape(-1, 1),
                                                    y_train
                                                    )
  y_pred = model.predict( np.array(X_test).reshape(-1, 1) )
  MSE[K] = mse( y_test, y_pred )

MSE
```

```python
MSE = pd.Series(MSE)
sns.lineplot(MSE)
```

```python
MSE[ MSE == MSE.min() ]
```

```python
#. K = 3 version of the KNN Regressor
model = KNeighborsRegressor( n_neighbors = 3 ).fit(
                                                  np.array(X_train).reshape(-1, 1),
                                                  y_train
                                                  )
y_pred = model.predict( np.array(X_test).reshape(-1, 1) )

#. Plot the test set performance of the KNN regressor
sns.scatterplot(
x = X_test,
y = y_test,
alpha = .1,
label = 'Original test set responses'
)
sns.lineplot(
x = X,
y = model.predict( np.array(X).reshape(-1, 1) ),
label = 'KNN predictions'
)
sns.lineplot(
x = X,
y = f(X),
label = 'Regression function',
alpha = 0.3
)
```
